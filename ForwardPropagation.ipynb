forward Propagation

```python
import numpy as np
#input
x = np.array([2.0])

# parameters
w = np.array([0.5])
b = 1.0

# forward propagation
z = np.dot(w,x) + b
y_hat = z # not activation (linear)

print("z:", z)
print("prediction:", y_hat)
```

```python
# with activation
def relu(z):
  return np.maximum(0,z)

z = np.dot(w,x)+b
a = relu(z)

print("Activated output:", a)
```

```python
# multilayer

#inputs
X = np.array([[1.0, 2.0]])

# weights and biases
W1 = np.array([
    [0.1,0.2],
    [0.3,0.4]
])
b1 = np.array([0.1,0.1])

W2 = np.array([
    [0.5],[0.6]
])
b2 = np.array([0.2])

# forward propagation
# layer 1
z1 = np.dot(X,W1) + b1
a1 = np.maximum(0,z1) # ReLU
# layer 2
z2 = np.dot(a1,W2) + b2
y_hat = z2       # Linear output

print("Prediction:", y_hat)
```

```python
# loss function
y = np.array([[10.0]])

loss = np.mean(np.square(y_hat - y))

print("Loss:", loss)
```

Backward Propagation

```python
# forward
x = np.array([2.0])
y = np.array([10.0])
w= np.array([0.5])
b =1.0
lr = 0.01

y_hat = np.dot(w,x) +b
loss = (y-y_hat) ** 2

# Backward
dL_dyhat = -2*(y-y_hat)
dL_dw = dL_dyhat * x
dL_db = dL_dyhat

# Update
w = w - lr * dL_dw
b = b - lr * dL_db

print("Updated weight:", w)
print("Updated bias:", b)
print("Updated loss:", loss)
```

```python
for epoch in range(10):
  # forward
  y_hat = np.dot(w,x) +b
  loss = (y-y_hat) ** 2

  # backward
  dL_dyhat = -2*(y-y_hat)
  dL_dw = dL_dyhat * x
  dL_db = dL_dyhat

  # update
  w = w - lr * dL_dw
  b = b - lr * dL_db

  print("Epoch:", epoch+1)
  print("Updated weight:", w)
  print("Updated bias:", b)
  print("Updated loss:", loss)
  print()
```
