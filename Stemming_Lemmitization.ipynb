```python
import nltk
import spacy
```

```python
from nltk import stem
from nltk.stem import PorterStemmer

stemmer = PorterStemmer()
```

```python
words = ["eating", "eats", "ate", "adjustable", "rafting", "ability", "meeting"]

for word in words:
    print(word, "|", stemmer.stem(word))
```

```python
nlp = spacy.load("en_core_web_sm")
# Dictionary

doc = nlp("eating eats ate adjustable rafting ability meeting better")
for token in doc:
    print(token, "|", token.lemma_)
```

Stop words Removal

```python
import nltk
nltk.download('punkt')
nltk.download('stopwords')
```

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

text = "The students are studying NLP and good to go is Friday available"
# Tokenize
tokens = word_tokenize(text)

# Load stop words
stop_words = set(stopwords.words("english"))

# Remove stop words
filtered_tokens = [token for token in tokens if token.lower() not in stop_words]

print("Original tokens:", tokens)
print("After stop-word removal:", filtered_tokens)
```

```python
custom_stopwords={"the", "is", "are", "in","go"}

filtered = [w for w in tokens if w.lower() not in custom_stopwords]
print(filtered)
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

text = "Once upon a time, there was a little bunny who loved to hop in the green forest. The bunny would hop and hop all day, but at night, he felt very sleepy and quiet. He made a soft bed under a big tree and looked at the bright moon in the sky. Soon, the little bunny closed his eyes and dreamed about fun adventures he would have the next day."

# Load NLTK stop words
nltk_stopwords = set(stopwords.words('english'))

# Remove stop words
filtered_words = [word for word in words if word not in nltk_stopwords]

print(filtered_words)
```

```python
# Custom stop words
custom_stopwords = {"bunny", "little", "hop", "forest", "day", "night", "tree", "moon", ",", "."}

# Combine both
all_stopwords = nltk_stopwords.union(custom_stopwords)

# Tokenize
words = word_tokenize(text.lower())

# Remove stop words
filtered_words = [word for word in words if word not in all_stopwords]

print(filtered_words)
```
